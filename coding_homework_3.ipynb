```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
# I am importing all of the things we need to do all of the actions later in the notebook.
```

```python
url1 = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv'
v_rest_sst = np.loadtxt(url1)
# I am importing each raw data set as a numpy array!
url2= 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv'
v_rest_pvalb = np.loadtxt(url2)
```

```python
print("Mean v_rest_sst:", np.mean(v_rest_sst))
print("Mean v_rest_pvalb:", np.mean(v_rest_pvalb))
#I am printing both of the means of the sst and pvalb neurons seperatly so that you can see them at the top.
# I also added lables so that when the cell plays I can see what is what.

t_stat, p = stats.ttest_ind(v_rest_sst,v_rest_pvalb)
#I am running a t-test defining p value as p to use in my if statement.

if p < 0.05:
    fig, ax = plt.subplots(1, 2)   # 1 row, 2 columns

    ax[0].hist(v_rest_sst)
    ax[0].set_title("v_rest_sst")
    ax[0].set_ylabel('Marker')
    ax[0].set_xlabel('Resting Membrane Potential')

    ax[1].hist(v_rest_pvalb)
    ax[1].set_title("v_rest_pvalb")
    ax[1].set_ylabel('Marker')
    ax[1].set_xlabel('Resting Membrane Potential')

    plt.show()
#I created an if statement that will only plot the histograms of sst and pvalb if the p value is less than 0.05
# I labled each of the histograms and put them into their own grid togehter.

print('p-value:', p)
#I printed the p value just to see what it ended up being. It will be at the end.
```

```python
def vrestHist():
  """
  This function will print the mean of sst and pvalb, the p-value of a t test between sst and pvalb, and a histogram for both sst and pvalb.
  """

    print("Mean v_rest_sst:", np.mean(v_rest_sst))
    print("Mean v_rest_pvalb:", np.mean(v_rest_pvalb))


    t_stat, p = stats.ttest_ind(v_rest_sst, v_rest_pvalb)
    print("p-value:", p)

    if p < 0.05:
        fig, ax = plt.subplots(1, 2)

        ax[0].hist(v_rest_sst)
        ax[0].set_title("v_rest_sst")

        ax[1].hist(v_rest_pvalb)
        ax[1].set_title("v_rest_pvalb")

    plt.show()
#This code cell does the exact same thing as the last code cell but defines all the actions into one function called vrestHist
#I am still printing the means of sst and pvalb and the p value with an if statement that will print histograms if the means are statistically significantly different.
```
